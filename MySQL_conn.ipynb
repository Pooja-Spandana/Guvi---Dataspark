```python
# importing libraries

import pandas as pd
import mysql.connector
from sqlalchemy import create_engine
```

```python
# Read the sales file

sales = pd.read_csv(r"Clean datasets\Sales_C.csv")
sales.head()
```

```python
# Read the stores file

stores = pd.read_csv(r"Clean datasets\Stores_C.csv")
stores.head()
```

```python
# Read the products file

products = pd.read_csv(r"Clean datasets\Products_C.csv")
products.head()
```

```python
# Read the customers file

customers = pd.read_csv(r"Clean datasets\Customers_C.csv")
customers.head()
```

```python
# Read the exchange_rates file

exchange_rates = pd.read_csv(r"Clean datasets\Exchange_Rates_C.csv")
exchange_rates.head()
```

```python
# Read the Global_electronics_data file

Global_electronics_data = pd.read_csv(r"Clean datasets\Global_electronics_data_C.csv")
Global_electronics_data.head()
```

```python
# SQL connection

client = mysql.connector.connect(
    host="localhost", 
    user="root", 
    password="root", 
    auth_plugin="mysql_native_password"
)

print("Connection successful!")
cursor = client.cursor()

# SQL query to create a database

cursor.execute("CREATE DATABASE IF NOT EXISTS DATASPARK")
print("Database created successfully!")
```

```python
# Create an SQLAlchemy engine for MySQL
engine = create_engine("mysql+mysqlconnector://root:root@localhost/DATASPARK?auth_plugin=mysql_native_password")

# Writing all DataFrames to MySQL tables

sales.to_sql("sales", con=engine, if_exists="replace", index=False)
stores.to_sql("stores", con=engine, if_exists="replace", index=False)
products.to_sql("products", con=engine, if_exists="replace", index=False)
customers.to_sql("customers", con=engine, if_exists="replace", index=False)
exchange_rates.to_sql("exchange_rates", con=engine, if_exists="replace", index=False)
Global_electronics_data.to_sql("global_electronics", con=engine, if_exists="replace", index=False)

print("Tables Created successfully")
```
